```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("breast-cancer-dataset.csv")
df.head()
```

```python
if 'id' in df.columns:
    df = df.drop('id', axis=1)

df['Diagnosis Result'] = df['Diagnosis Result'].map({'Benign': 0, 'Malignant': 1})

X = df.drop('Diagnosis Result', axis=1).values
y = df['Diagnosis Result'].values
```

```python
X_columns = df.drop('Diagnosis Result', axis=1).columns
X_df = pd.DataFrame(X, columns=X_columns)

for col in X_df.columns:

    try:
        X_df[col] = pd.to_numeric(X_df[col])
    except ValueError:
        pass

categorical_cols = X_df.select_dtypes(include=['object']).columns

if len(categorical_cols) > 0:
    X_processed_df = pd.get_dummies(X_df, columns=categorical_cols, drop_first=True)
else:
    X_processed_df = X_df


X = X_processed_df.values.astype(float)


X = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))
```

```python
split = int(0.8 * len(X))

X_train = X[:split]
y_train = y[:split]

X_test = X[split:]
y_test = y[split:]
```

```python
class Perceptron:

    def __init__(self, lr=0.01, epochs=100):
        self.lr = lr
        self.epochs = epochs
        self.weights = None
        self.bias = None

    def activation(self, x):
        return np.where(x >= 0, 1, 0)

    def fit(self, X, y):
        n_samples, n_features = X.shape

        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.epochs):
            for i in range(n_samples):
                linear_output = np.dot(X[i], self.weights) + self.bias
                y_pred = self.activation(linear_output)

                update = self.lr * (y[i] - y_pred)

                self.weights += update * X[i]
                self.bias += update

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        return self.activation(linear_output)
```

```python
model = Perceptron(lr=0.01, epochs=100)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = np.sum(y_pred == y_test) / len(y_test)
print("Accuracy:", accuracy * 100)
```
